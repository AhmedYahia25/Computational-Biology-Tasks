```python
import pandas as pd
```

```python
df=pd.read_csv('HomoSapiens_binary_hq.txt',nrows=100,sep="\t",usecols=['Gene_A','Gene_B'])
```

```python
df 
```

```python
#must specify which gene is source and destination
Bio_Graph=nx.from_pandas_edgelist(df,'Gene_A','Gene_B',create_using=nx.DiGraph())
```

```python
nx.draw(Bio_Graph,with_labels=True,node_size=1000)
```

```python
de=dict(Bio_Graph.degree())
```

```python
sorted(de.items())
```

```python
sorted(de.items(),key=lambda t:t[1])
```

```python
import matplotlib.pyplot as plt
```
